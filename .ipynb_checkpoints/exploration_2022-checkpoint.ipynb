```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des données

circonscriptions_2022 = pd.read_csv("data/2022/circonscriptions.csv", encoding="ISO-8859-1", sep=";")
votes_candidats_2022 = pd.read_csv("data/2022/candidats.csv", encoding="ISO-8859-1", sep=";")
```

```python
# Affichage des titres de chacune des colonnes

print(circonscriptions_2022.columns)
print(votes_candidats_2022.columns)
```

```python
# Affichage des cinq premières lignes, histoire de voir de quoi les données ont l'air

circonscriptions_2022.head()
```

```python
# La colonne de mise à jour au format ISO 8601 n'est pas utile pour le travail à venir, tout comme le booléen pour les résultats finaux

circonscriptions_2022 = circonscriptions_2022.drop(columns=['Dernière date de mise à jour au format ISO 8601', 'Résultats finaux'])

# Afin d'alléger le reste de la prodécure, je renomme une des colonnes

votes_candidats_2022 = votes_candidats_2022.rename(columns={'Abréviation du parti politique': 'Parti politique'})
```

```python
# Vérification de l'intégrité des données

# Il devrait normalement y avoir 125 lignes, une par circonscription électorale au Québec, et les colonnes Numéro et Nom de la circonscriptions devraient êtres uniques

print("Nombre de circonscriptions(lignes) dans le dataset : ", len(circonscriptions_2022))
numero_unique = circonscriptions_2022['Numéro de la circonscription'].is_unique
nom_unique = circonscriptions_2022['Nom de la circonscription'].is_unique

print("Les numéros de circonscriptions sont-ils uniques? ", numero_unique)
print("Les noms de circonscriptions sont-ils uniques? ", nom_unique)
```

```python
# Quelles sont les circonscriptions avec le plus haut et le plus faible taux de participation?
# On commence d'abord par isoler les colonnes utiles, ici le nom de la circonscription et le taux de participation

taux_participation_2022 = (circonscriptions_2022.iloc[:,[1,10]]).sort_values(by='Taux de participation', ascending=False)

# Ensuite on divise en quintile, en ajoutant +1 pour avoir des quintiles de 1 à 5 plutôt que de 0 à 4
taux_participation_2022['quintile'] = pd.qcut(taux_participation_2022['Taux de participation'], q=5, labels=False) + 1

# On affiche toutes les circonscriptions du quintile avec le taux de participation le plus élevé
print("Les circonscriptions se trouvant dans le quintile avec le plus haut taux de participation sont : ")
print(taux_participation_2022[taux_participation_2022['quintile'] == 5])
print("-------------------------------------------------------------------------------------------------")

# Ensuite celle dans le dernier quintile
print("Les circonscriptions se trouvant dans le quintile avec le plus bas taux de participation sont : ")
print(taux_participation_2022[taux_participation_2022['quintile'] == 1])
```

```python
# Trouver candidat vainqueur pour chaque circonscription (votes en avance plus grand que zéro)

vainqueurs_2022 = votes_candidats_2022[votes_candidats_2022['Nombre de votes en avance'] > 0]
```

```python
# Trouver le résultats des élection en terme de notre de candidats élus

candidats_elus_parparti_2022 = vainqueurs_2022['Parti politique'].value_counts()

print(candidats_elus_parparti_2022.to_string())
```

```python
# Maintenant le résultat des élections, en nombre de votes par parti

total_votes_parparti_2022 = votes_candidats_2022.groupby('Parti politique')['Nombre total de votes'].sum()
print(total_votes_parparti_2022)
total_votes_parparti_2022.plot(kind='bar', figsize=(10, 6))
plt.title("Réparitions des votes par parti")
plt.ylabel("en millions de votes")
plt.show()
```

```python
# Intéressant à regarde:, la différence entre le résultat du vote populaire versus les sièges obtenus.
# Le vote populaire étant le suffrage effectivement reçu.  Cela nous donnerait une idée de ce que serait la répartition
# des partis si on changerait le mode électoral pour quelque chose de plus portionnel.  J'en profite pour signaler une
# initiative pancanadienne qui milite pour ce genre de chose : https://www.fairvote.ca/fr/

total_sieges_2022 = candidats_elus_parparti_2022.values.sum()
proportion_sieges_2022 = candidats_elus_parparti_2022 / total_sieges_2022 * 100

print("Proportion des sièges obtenus (en pourcentage) : ")
print(proportion_sieges_2022)

print("Proportion des votes reçus (en pourcentage : ")
proportion_vote_2022 = total_votes_parparti_2022 / total_votes_parparti_2022.sum() * 100
print(proportion_vote_2022)
```

```python
# Avec un vote proportionnel donc, la composition de l'Assemblée Nationale serait très différente, le gouvernement élu
# serait notamment minoritaire, au lieu d'avoir l'importante majorité actuelle.  Les distorsions de ce système sont assez
# claires à cette élection.
```

```python
# Trouver le candidat élu avec le moins de votes

elu_moinsdevotes_2022 = vainqueurs_2022.loc[vainqueurs_2022['Nombre total de votes'].idxmin()]
print(elu_moinsdevotes_2022)
```

```python
# Tâchons de comparer maintenant tous les candidats vainqueurs par le nombre de votes obtenus

vainqueurs_triespartotalvotes_2022 = vainqueurs_2022[['Prénom', 'Nom', 'Nombre total de votes']].sort_values(by='Nombre total de votes', ascending=False)
vainqueurs_triespartotalvotes_2022['nom complet'] = vainqueurs_triespartotalvotes_2022['Prénom'] + ' ' + vainqueurs_triespartotalvotes_2022['Nom']
```

```python
plt.figure(figsize=(5, 20))
plt.barh(vainqueurs_triespartotalvotes_2022['nom complet'], vainqueurs_triespartotalvotes_2022['Nombre total de votes'])
plt.show()
```

```python
# Ceci dit étant donné que les circonscriptions n'ont pas toute le même nombre d'électeurs. Le résultat précédent
# n'est pas aussi intéressant qu'il en a l'air, surtout si on ajoute le fait que le taux de participation est vraiment inégal.
```

```python
# Regardons maintenant le taux de vote pour le candidat gagnant dans chacune des circonscriptions et le nombre de votes en 
# avance pour le vainqueur dans chacune d'elles.

comparaison_avance_taux_partaux_2022 = vainqueurs_2022[['Numéro de la circonscription', 'Nombre de votes en avance', 'Taux de vote']].sort_values(by='Taux de vote')
comparaison_avance_taux_paravance_2022 = vainqueurs_2022[['Numéro de la circonscription', 'Nombre de votes en avance', 'Taux de vote']].sort_values(by='Nombre de votes en avance')

print("Les cinq circonscriptions avec le plus bas taux de vote.")

print(comparaison_avance_taux_partaux_2022.head())

# Il semble y avoir une tendance à avoir moins d'avance lorsque le taux de vote est plus bas, ce qui a du sens quand on y 
# pense.

print("\nLes cinq circonscriptions avec le plus haut taux de vote.")

print(comparaison_avance_taux_partaux_2022.tail())

print("\nLes cinq circonscriptions avec l'avance la plus basse.")

print(comparaison_avance_taux_paravance_2022.head())

print("\nLes cinq circonscriptions avec les plus hautes avances.")

print(comparaison_avance_taux_paravance_2022.tail())
```

```python
# Jetons un oeil sur les moyennes et médianes.

print("Taux de vote moyen : ", comparaison_avance_taux_partaux_2022['Taux de vote'].mean())
print("Taux de vote médian : ", comparaison_avance_taux_partaux_2022['Taux de vote'].median())
print("Avance moyenne : ", comparaison_avance_taux_paravance_2022['Nombre de votes en avance'].mean())
print("Avance médiane : ", comparaison_avance_taux_paravance_2022['Nombre de votes en avance'].median())
```

```python
# On note que la moyenne ressemble à la médiane, ce qui laisse croire qu'on a quelque chose pas trop loin d'une courbe 
# normale. Ceci dit on regarde ici le taux de vote pour le vainqueur dans une circonscription

plt.figure(figsize=(10, 6))
sns.histplot(vainqueurs_2022['Taux de vote'], kde=True, stat="density", bins=20, color='skyblue')

# Ajouter une ligne pour la moyenne
moyenne = vainqueurs_2022['Taux de vote'].mean()
plt.axvline(moyenne, color='red', linestyle='--', label=f'Moyenne: {moyenne:.2f}')
mediane = vainqueurs_2022['Taux de vote'].median()
plt.axvline(mediane, color='purple', linestyle='--', label=f'Médiane: {mediane:.2f}')

plt.title("Distribution du nombre de votes par circonscription")
plt.xlabel("Nombre de votes")
plt.ylabel("Densité")
plt.legend()
plt.show()
```

```python
# C'est somme toute assez proche d'une courbe normale.
```

```python
# Autre chose intéressante à explorer : la taille des circonscriptions.

plt.figure(figsize=(10, 6))
sns.histplot(circonscriptions_2022['Nombre d\'électeurs inscrits'], kde=True, stat="density", bins=20, color='skyblue')

# Ajouter une ligne pour la moyenne
moyenne = circonscriptions_2022['Nombre d\'électeurs inscrits'].mean()
plt.axvline(moyenne, color='red', linestyle='--', label=f'Moyenne: {moyenne:.2f}')
mediane = circonscriptions_2022['Nombre d\'électeurs inscrits'].median()
plt.axvline(mediane, color='purple', linestyle='--', label=f'Médiane: {mediane:.2f}')

plt.title("Taille des circonscriptions, en utilisant le nombre d'électeurs inscrits")
plt.xlabel("Nombre d'électeurs inscrits")
plt.ylabel("Densité")
plt.legend()
plt.show()
```

```python
# Parmis les critères qu'utilisent la Commission de la représentation électorale il y a entre autres l'aspect
# du nombre d'électeurs qui doit être le plus équitable possible.  Ce n'est pas surprenant donc que la majorité
# des circonscriptions se retrouvent à plus ou moins 10 000 électeurs de la moyenne.
# Mais qu'elles sont les circonscriptions sous la barre des 40 000 électeurs?

circonscriptions_moins_40000_2022 = circonscriptions_2022[circonscriptions_2022['Nombre d\'électeurs inscrits'] < 40000].sort_values(by='Nombre d\'électeurs inscrits', ascending=False)
print(circonscriptions_moins_40000_2022[['Nom de la circonscription', 'Nombre d\'électeurs inscrits']])
```
