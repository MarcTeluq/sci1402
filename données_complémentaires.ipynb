```python
import pandas as pd
```

```python
# "dc" est une abbréviation de "données complètes 2021"
dc_original = pd.read_csv("data/statistiques-recensement-2021-CEP.csv", encoding="ISO-8859-1", sep=";")
```

```python
# Première étape : inverser les colonnes et les lignes
dc_original = dc_original.set_index('Circonscription')
dc = dc_original.transpose()
dc.head()
```

```python
# Supprimer les espaces et remplacer les virgules par des points dans toutes les colonnes sauf "Circonscription"
colonnes_a_convertir = [col for col in dc.columns if col != 'Circonscription']

for colonne in colonnes_a_convertir:
    dc[colonne] = dc[colonne].astype(str).str.replace(' ', '').str.replace(',', '.')
    dc[colonne] = pd.to_numeric(dc[colonne], errors='coerce')
```
